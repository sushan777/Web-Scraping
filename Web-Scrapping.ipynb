```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Firefox(executable_path=r"C:\Users\sthrita\Desktop\ds\geckodriver.exe")
driver.get("http://nepalstock.com.np/company/transactions/134")
```

```python
def get_this_page_data(driver):
    table = driver.find_element_by_css_selector('.table.my-table')
    table_html=table.get_attribute('innerHTML')
    table_html = '<html><head></head><body><table>'+table_html.replace('\n','')+'</table></body></html>'
    soup = BeautifulSoup(table_html)
    data=[]
    for row in soup.find_all('tr')[2:-1]:
        transaction=[]
        for d in row.find_all('td'):
            transaction.append(d.text)
        data.append(transaction)
    return data
```

```python
def get_next_page(driver):
    try:
        next_page = driver.find_element_by_xpath('//a[@title="Next Page"]')
        next_page.click()
        return True
    except:
        return False
    
```

```python
def get_all_data(driver):
    data = get_this_page_data(driver)
    while get_next_page(driver):
        data.extend(get_this_page_data(driver))
    return data
```
